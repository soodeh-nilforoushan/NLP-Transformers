```python
from transformers import pipeline
import numpy as np
import pandas as pd
import textwrap
from pprint import pprint
```

```python
df=pd.read_csv('bbc_text_cls.csv')
```

```python
df.head()
```

```python
#pick an article random from business class
# we set random_set to 42 so that we can replicate the result
doc=df[df.labels=='business']['text'].sample(random_state=42)
print(doc)
```

```python
# the text is so longand go off the page so we use this function
def wrap(x):
  return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)
```

```python
print(wrap(doc.iloc[0]))
```

```python
#as defualt it use distilbart. which is a kind of seq2seq. Such models useful for tasks like language translation, where the output seq length is not equall to input seq length
# this is abstractive summarization since the model should generate a summary based on the representation of input
summarize=pipeline("summarization")
```

```python
#we need to split the first line from other, because usually first line shows the summary of text
#and because we do not wanna cheat, we only work with the second line of text
summarize(doc.iloc[0].split("\n",1)[1])
```

```python
#just print the above text in sepicific format
def print_summary(doc):
  result =summarize(doc.iloc[0].split("\n",1)[1])
  print(wrap(result[0]['summary_text']))
```

```python
print_summary(doc)
#it is a good summary and the sentences are not exactly the same as input
```

```python
#select another random article
doc=df[df.labels=='entertainment']['text'].sample(random_state=123)
print(wrap(doc.iloc[0]))
```
